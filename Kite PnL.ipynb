```python
import pandas as pd
```

```python
df= pd.read_csv('orders.csv')
```

# Q.1 Different types of charges for Individual trade

```python
df
```

```python
com_status= df[(df['Product']=='MIS')]
com_status
```

```python
df1=com_status
df1.reset_index(inplace = True )
```

```python
df1
```

```python
df1['Qty.']=df1['Qty.'].str.split('/').str[0]
```

```python
df1
```

```python
# check the datatypes
df1.dtypes
```

```python
#converting dataypes from float to int 
df1['Qty.']=df1['Qty.'].astype('int')
```

```python
# calculating Turnover
df1['Turnover']=df1['Qty.']*df1['Avg. price']
```

```python
#converting dataypes from float to int 
df1['Turnover']=df1['Turnover'].astype('int')
```

```python
df1
```

# Calculating Brokerage

```python
#df1['Brokerage']=df1['Status']='COMPLETED'=20 

#finding Brokerage charges
df1.loc[df1['Status'] == 'COMPLETE', 'Brokerage'] = 20
df1
```

```python
#dropping/ Replacing NAN values with 0
df1['Brokerage'].fillna(0, inplace= True)

#converting float to int
df1['Brokerage']= df1['Brokerage'].astype('int')
```

```python
df1 
```

# Calculating STT/CTT

```python
#df1['STT/CTT'] = (0.025* df1['Turnover'] )/100
#df1.loc[df1['Status'] == 'COMPLETE', 'Brokerage'] = 20

df1.loc[df1['Type']=='SELL','STT/CTT']= round(0.025* df1['Turnover'])/100
```

```python
#dropping NAN Values
df1['STT/CTT'].fillna(0,inplace = True)
```

# Calculating Transaction/ ETC

```python
df1['ETC']=(0.00325*df1['Turnover'])/100
```

```python
df1
```

# Calculating SEBI 

```python
s= round(df1['Turnover']/1000000,2)
s1=(18*s)/100+s

df1['SEBI']=s1 


#(18*df1['Turnover']/10000000)/100
#df1.loc[df1['Type']=='SELL','STT/CTT']= round(0.025* df1['Turnover'])/100
#df1['Turnover']/1000000
```

```python
df1
```

# Calculate GST

```python
df1['GST']=round(18*(df1['Brokerage']+df1['ETC']+df1['SEBI'])/100, 2)
```

# Calculate Stamp_Charges 

```python
df1['Stamp_Charges'] = round(0.003*df1['Turnover']/100,2)
```

# Calculating Total Charges

```python
#df1[['Brokerage','STT/CTT','ETC','SEBI','GST','Stamp_Charges']].iloc[0].sum()

df1['Total Charges'] = round(df1[['Brokerage','STT/CTT','ETC','SEBI','GST','Stamp_Charges']].sum(axis=1),2)
```
